# Exploratory Data Analysis (EDA)

## Objective

The objective of this analysis is to explore patient demographics, hospital admission characteristics, and clinical factors associated with hospital readmissions. The findings will help identify trends and potential risk factors that can support better healthcare decision-making.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


pd.set_option("display.max_columns", None)        #diplay all 45 cols
sns.set_theme(style="whitegrid")                  #change default apperance to white bg and grid lines
plt.rcParams["figure.figsize"] = (8,5)            #set all figure size to 8,5



db=pd.read_csv('/content/diabetic_data_cleaned.csv')
db.head(10)
```

```python
db.shape
```

```python
db.info()
```

```python
db.isnull().sum()
```

```python
num_cols=db[['admission_type_id','discharge_disposition_id','admission_source_id','time_in_hospital','num_lab_procedures','num_procedures','num_medications','number_diagnoses']]
```

```python
#distribution of numeric cols using histogram

for col in num_cols:
  plt.figure()
  sns.histplot(db[col],kde=True, color='skyblue')
  plt.title(f'Distribution of {col}')
  plt.show()

```

```python
for col in num_cols:

    plt.figure()

    sns.boxplot(x=db[col])

    plt.title(f"Boxplot of {col}")

    plt.show()
```

```python
cat_cols=db[['race','gender','age','change','readmitted','diabetesMed']]
for col in cat_cols:
  plt.figure()
  sns.boxplot(x=db[col])
  plt.title(f'Distribution of {col}')
  plt.show()
```

```python
for col in cat_cols:
  plt.figure()
  sns.countplot(x=db[col])
  plt.title(f'Distribution of {col}')
  plt.show()
```

```python
db['readmitted'].value_counts(normalize=True)*100
```

```python
#Readmission by age
plt.figure()
sns.countplot(x=db['age'], hue=db['readmitted'])
plt.title('Readmission by Age')
plt.show()
```

key info: Patients within [70-80]age group were readmitted within 30 days of discharge.



```python
#Readmission by Gender
plt.figure()
sns.countplot(x=db['gender'],hue=db['readmitted'])
plt.title('Readmission by Gender')
plt.show()
```

key info: women tends to have more readmission within 30 days of discharge.

```python
#Readmission by Race
plt.figure()
sns.countplot(x=db['race'],hue=db['readmitted'])
plt.title('Readmission by Race')
plt.show()
```

key info: People belonging to caucasian race have high proportion of readmission

```python
#time in hospital vs readmission
plt.figure()
sns.boxplot(data=db, x='readmitted',y='time_in_hospital')
plt.title('Hospital Stay by readmssion status')
plt.show()
```

key finding: The distribution of hospital stay duration is similar across all readmission groups. The median stay is approximately 4 days for patients who were not readmitted, readmitted within 30 days, and readmitted after 30 days. Although a few patients experienced longer hospital stays (13–14 days), these appear as outliers in all groups

```python
#medications vs readmissions
plt.figure()
sns.boxplot(data=db,x='readmitted',y='num_medications')
plt.title('Number of Medications by Readmission Status')
plt.show()
```

key info: Patients who were readmitted within 30 days  tend to receive slightly more medications than the other two groups.
